```python
from transformers import AutoTokenizer, AutoModelForCausalLM

model_name = "autumnal05/finetune_normal_tissue_4k"

#Loading tokenizer and model
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name, device_map="auto", load_in_8bit=True)#loading the model in 8 bit precision due to colab memory constraints
```

```python
input_text = "What is the expression profile for TSPAN6 in the tonsil?"
inputs = tokenizer(input_text, return_tensors="pt")

#Generating response with a limited number of tokens
outputs = model.generate(**inputs, max_new_tokens=30)
response = tokenizer.decode(outputs[0], skip_special_tokens=True)

print(response)
```
